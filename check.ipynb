```python
import pandas as pd
```

```python
stripped_june = pd.read_csv('siboehm_new_eval_stripped.csv')
```

```python
data = pd.read_csv(open('generation/LEETCODE_EASY_EVAL.csv'))
name = "leetcode"
non_stripped_june = data[data['model'].str.contains("gpt-4-0613")].copy()
```

```python
def direct_usable(datapoint):
    if ('Accepted' in datapoint['Code_Submit']):
        return 1
    if ('Error' in datapoint['Code_Submit']):
        return 0
    if ('Wrong' in datapoint['Code_Submit']):
        return 0    
    print("Error! No Accept or Error")
    return 0
```

```python
stripped_june['Directly Usable'] = stripped_june.apply(direct_usable,axis=1)
non_stripped_june['Directly Usable'] = non_stripped_june.apply(direct_usable,axis=1)
```

```python
stripped_june["Directly Usable"].value_counts()
```

```python
non_stripped_june["Directly Usable"].value_counts()
```

```python
data = pd.read_csv(open('generation/LEETCODE_EASY_EVAL.csv'))
name = "leetcode"
march_gpt = data[data['model'].str.contains("gpt-4-0314")]
```

```python
march_gpt["Directly Usable"].value_counts()
```

```python
# now we combine the three value_counts
temp = pd.DataFrame()
temp['june_fixed'] = stripped_june["Directly Usable"].value_counts()
temp['june_orig'] = non_stripped_june["Directly Usable"].value_counts()
temp['march_orig'] = march_gpt["Directly Usable"].value_counts()
# turn into false and true
temp = temp.rename(index={0: 'False', 1: 'True'})
# rename index
temp = temp.rename_axis('Leetcode accept')
temp
```
